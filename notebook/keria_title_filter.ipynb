```python
url = 'https://www.kats.go.kr/content.do?cmsid=239&page=1'
def get_table(url):
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    table = pd.read_html(str(soup.select('body > div.sub_content > ul > li > ul > li.sub_right > div.subpage_con > table')[0]))
    return table[0]
```

```python
get_table(url)
```

```python
tables = [get_table(url)]
for i in range(1, 200):
    url = f'https://www.kats.go.kr/content.do?cmsid=239&page={i}'
    table = get_table(url)
    tables.append(table)
    if not i%10:
        print('page', i, 'done')
```

```python
all_tables = pd.concat(tables)
for line in all_tables['제목']:
    print(line)
    break
```
